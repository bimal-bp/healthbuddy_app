```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Load Data
df = data.copy()

# Preprocessing foodname
df['foodname'] = df['foodname'].str.lower().str.replace('[^a-z\s]', '', regex=True)

# Features and Target
X = df['foodname']
y = df[['Vitamin A', 'Vitamin B12', 'Vitamin B6', 'Vitamin C', 'Vitamin E', 'Vitamin K']]

# Convert text to numerical features using TF-IDF
vectorizer = TfidfVectorizer(max_features=500)  # Adjust max_features as needed
X_tfidf = vectorizer.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Model
model = MultiOutputRegressor(RandomForestRegressor(random_state=42))

# Train
model.fit(X_train, y_train)

# Evaluate
score = model.score(X_test, y_test)
print(f"Model R^2 Score: {score:.2f}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
import re

# Example data setup


# Define thresholds for deficiency (e.g., threshold for deficiency is set arbitrarily to below 5 for each vitamin)
deficiency_thresholds = {
    'Vitamin A': 5,
    'Vitamin B12': 0.5,
    'Vitamin B6': 0.2,
    'Vitamin C': 1.0,
    'Vitamin E': 0.4,
    'Vitamin K': 2.0
}

# Convert continuous vitamin values to binary (0 = sufficient, 1 = deficient)
for vitamin, threshold in deficiency_thresholds.items():
    data[vitamin + '_deficiency'] = (data[vitamin] < threshold).astype(int)

# Define features (X) and target (y)
X = data['foodname']  # Feature: food names
y = data[[vitamin + '_deficiency' for vitamin in deficiency_thresholds.keys()]]  # Target: deficiency labels

# Convert text to numerical features using BoW
vectorizer = CountVectorizer(max_features=500)
X_bow = vectorizer.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_bow, y, test_size=0.2, random_state=42)

# Model (Multi-output classification)
model = MultiOutputClassifier(RandomForestClassifier(random_state=42))

# Train the model
model.fit(X_train, y_train)

# Evaluate the model
score = model.score(X_test, y_test)
print(f"Model Accuracy: {score * 100:.2f}%")

# Predict deficiencies for new food items
new_foodnames = ['Beef and noodles with mushroom sauce', 'Pepperoni, reduced fat', 'Corned beef patty',
                 'Infant formula, ready-to-feed (Enfamil A.R.)', 'Frankfurter or hot dog sandwich, with meatless...']

# Preprocess and transform the new food names using the fitted vectorizer
new_foodnames_processed = [re.sub(r'[^a-z\s]', '', name.lower()) for name in new_foodnames]  # Clean and lower case
new_X_bow = vectorizer.transform(new_foodnames_processed)

# Make predictions
predictions = model.predict(new_X_bow)

# Define the vitamin names corresponding to the deficiencies
vitamin_names = ['Vitamin A', 'Vitamin B12', 'Vitamin B6', 'Vitamin C', 'Vitamin E', 'Vitamin K']

# For each food item, get the vitamins that are deficient (where the value is 1)
deficient_vitamins = []
for pred in predictions:
    deficient = [vitamin_names[i] for i in range(len(pred)) if pred[i] == 1]
    deficient_vitamins.append(deficient)

# Show the results
for i, food in enumerate(new_foodnames):
    print(f"Food: {food}")
    print(f"Deficient Vitamins: {', '.join(deficient_vitamins[i]) if deficient_vitamins[i] else 'None'}\n")
```

```python
import pickle
import pandas as pd
import re
import json
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
data=pd.read_csv('D:/healthbuddy app/food.csv')
# Define thresholds for deficiency (e.g., arbitrary thresholds)
deficiency_thresholds = {
    'Vitamin A': 5,
    'Vitamin B12': 0.5,
    'Vitamin B6': 0.2,
    'Vitamin C': 1.0,
    'Vitamin E': 0.4,
    'Vitamin K': 2.0
}

# Example DataFrame setup for testing purposes
data = pd.DataFrame({
    'foodname': ['Beef and noodles with mushroom sauce', 'Pepperoni, reduced fat', 'Corned beef patty'],
    'Vitamin A': [4, 1, 6],
    'Vitamin B12': [0.3, 0.5, 1.2],
    'Vitamin B6': [0.1, 0.3, 0.4],
    'Vitamin C': [0.9, 1.0, 0.2],
    'Vitamin E': [0.2, 0.5, 0.3],
    'Vitamin K': [1.5, 3.0, 0.8]
})

# Convert continuous vitamin values to binary (0 = sufficient, 1 = deficient)
for vitamin, threshold in deficiency_thresholds.items():
    data[vitamin + '_deficiency'] = (data[vitamin] < threshold).astype(int)

# Define features (X) and target (y)
X = data['foodname']  # Feature: food names
y = data[[vitamin + '_deficiency' for vitamin in deficiency_thresholds.keys()]]  # Target: deficiency labels

# Convert text to numerical features using BoW
vectorizer = CountVectorizer(max_features=500)
X_bow = vectorizer.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_bow, y, test_size=0.2, random_state=42)

# Model (Multi-output classification)
model = MultiOutputClassifier(RandomForestClassifier(random_state=42))

# Train the model
model.fit(X_train, y_train)

# Evaluate the model
score = model.score(X_test, y_test)
print(f"Model Accuracy: {score * 100:.2f}%")

# Model evaluation by individual vitamin
for vitamin in deficiency_thresholds.keys():
    vitamin_index = list(deficiency_thresholds.keys()).index(vitamin)
    vitamin_accuracy = (y_test.iloc[:, vitamin_index] == model.predict(X_test)[:, vitamin_index]).mean()
    print(f"Accuracy for {vitamin}: {vitamin_accuracy * 100:.2f}%")

# Save the trained model and vectorizer to pickle files
with open('model.pkl', 'wb') as model_file:
    pickle.dump((model, vectorizer), model_file)
print("Model and vectorizer saved as 'model.pkl'")

# Save additional configurations like deficiency thresholds and vitamin names
config = {
    "deficiency_thresholds": deficiency_thresholds,
    "vitamin_names": list(deficiency_thresholds.keys()),
    "model_description": "RandomForest-based multi-output classifier for vitamin deficiencies"
}
with open('config.json', 'w') as config_file:
    json.dump(config, config_file, indent=4)
print("Configuration saved as 'config.json'")

# Predict deficiencies for new food items
with open('model.pkl', 'rb') as model_file:
    loaded_model, loaded_vectorizer = pickle.load(model_file)

new_foodnames = [
    'Beef and noodles with mushroom sauce',
    'Pepperoni, reduced fat',
    'Corned beef patty',
    'Infant formula, ready-to-feed (Enfamil A.R.)',
    'Frankfurter or hot dog sandwich, with meatless...'
]

# Preprocess and transform the new food names
new_foodnames_processed = [re.sub(r'[^a-z\s]', '', name.lower()) for name in new_foodnames]
new_X_bow = loaded_vectorizer.transform(new_foodnames_processed)

# Make predictions
predictions = loaded_model.predict(new_X_bow)

# Define the vitamin names corresponding to the deficiencies
vitamin_names = list(deficiency_thresholds.keys())

# For each food item, get the vitamins that are deficient
print("\nDeficiency Results:")
deficient_vitamins = []
for i, pred in enumerate(predictions):
    deficient = [vitamin_names[j] for j in range(len(pred)) if pred[j] == 1]
    deficient_vitamins.append(deficient)
    print(f"Food: {new_foodnames[i]}")
    print(f"Deficient Vitamins: {', '.join(deficient) if deficient else 'None'}")

# Identify the majority deficient vitamin
all_deficient_vitamins = [vitamin for sublist in deficient_vitamins for vitamin in sublist]
vitamin_count = Counter(all_deficient_vitamins)
if vitamin_count:
    majority_vitamin = vitamin_count.most_common(1)[0][0]
    print(f"\nMajority Deficient Vitamin: {majority_vitamin}")
else:
    print("\nNo vitamins are deficient in the provided food items.")
```
